```python
tasks = []

while True:

    print("\n--- TO DO LIST MENU ---")
    print("1. Add Task")
    print("2. View Task")
    print("3. Remove Task")
    print("4. Exit")

    choice = input("Enter your choice: ")

    if choice == '1':
        task = input("Enter new task: ")
        tasks.append(task)
        print("Task added successfully")

    elif choice == '2':
        if len(tasks) == 0:
            print("No tasks available")
        else:
            print("Your Tasks:")
            for t in tasks:
                print(t)

    elif choice == '3':
        task = input("Enter task to remove: ")
        if task in tasks:
            tasks.remove(task)
            print("Task removed")
        else:
            print("Task not found")

    elif choice == '4':
        print("Program ended")
        break

    else:
        print("Invalid choice")
```
